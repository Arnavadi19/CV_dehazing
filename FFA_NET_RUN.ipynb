```python
from ultralytics import YOLO
import os
import cv2
import json
from typing import Dict, List

def run_yolo_detection(
    image_folder: str,
    output_detection_folder: str,
    output_json_folder: str,
    model_name: str,
    json_filename: str = "detection_metrics.json",
    save_json: bool = True,
    save_detections: bool = True,
    max_images: int = None
) -> Dict[str, List]:

    metrics = {
        'filenames': [],
        'detections': []
    }

    model = YOLO(model_name)
    os.makedirs(output_detection_folder, exist_ok=True)
    os.makedirs(output_json_folder, exist_ok=True)

    processed_count = 0
    image_files = sorted([f for f in os.listdir(image_folder)
                         if f.lower().endswith(('.jpg', '.jpeg', '.png'))])

    if max_images is not None:
        image_files = image_files[:max_images]

    for image_file in image_files:
        image_path = os.path.join(image_folder, image_file)
        results = model(image_path)

        for result in results:
            image_metrics = {
                'classes': [],
                'names': [],
                'confidences': [],
                'boxes': []
            }

            if save_detections:
                output_path = os.path.join(output_detection_folder, f'detected_{image_file}')
                cv2.imwrite(output_path, result.plot())

            for box in result.boxes:
                image_metrics['classes'].append(int(box.cls))
                image_metrics['names'].append(model.names[int(box.cls)])
                image_metrics['confidences'].append(float(box.conf))
                image_metrics['boxes'].append(box.xyxy.tolist()[0])

            metrics['filenames'].append(image_file)
            metrics['detections'].append(image_metrics)

        processed_count += 1
        if processed_count % 50 == 0:
            print(f"{processed_count} images have been processed")
            break

    if save_json:
        json_path = os.path.join(output_json_folder, json_filename)
        with open(json_path, 'w') as f:
            json.dump(metrics, f, indent=2)
        print(f"Metrics saved to {json_path}")

    print(f"Processing complete. {processed_count} images processed.")
    return metrics
```

## Detection comparison

```python
import json
import numpy as np
from typing import Dict, List, Tuple

def load_metrics(json_path: str) -> Dict:
    with open(json_path) as f:
        metrics = json.load(f)

    required_keys = {'filenames', 'detections'}
    if not all(key in metrics for key in required_keys):
        raise ValueError(f"Invalid JSON structure. Missing required keys: {required_keys - set(metrics.keys())}")
    return metrics

def safe_mean(values: List[float]) -> float:
        return 0.0
    try:
        return float(np.mean(values))
    except:
        return 0.0

def compare_detections(metrics1: Dict, metrics2: Dict) -> Dict[str, Dict]:
    results = {}

    file_index1 = {name: i for i, name in enumerate(metrics1['filenames'])}
    file_index2 = {name: i for i, name in enumerate(metrics2['filenames'])}

    common_files = set(file_index1.keys()) & set(file_index2.keys())

    for filename in common_files:
        idx1 = file_index1[filename]
        idx2 = file_index2[filename]

        det1 = metrics1['detections'][idx1]
        det2 = metrics2['detections'][idx2]

        confidences1 = det1.get('confidences', []) or []
        confidences2 = det2.get('confidences', []) or []
        names1 = det1.get('names', []) or []
        names2 = det2.get('names', []) or []

        avg_conf1 = safe_mean(confidences1)
        avg_conf2 = safe_mean(confidences2)

        class_counts1 = {name: names1.count(name) for name in set(names1)}
        class_counts2 = {name: names2.count(name) for name in set(names2)}

        results[filename] = {
            'class_difference': {
                cls: class_counts1.get(cls, 0) - class_counts2.get(cls, 0)
                for cls in set(class_counts1) | set(class_counts2)
            },
            'confidence_diff': avg_conf1 - avg_conf2,
            'detection_count_diff': len(names1) - len(names2),
            'details': {
                'set1': {
                    'classes': names1,
                    'confidences': [float(c) for c in confidences1]
                },
                'set2': {
                    'classes': names2,
                    'confidences': [float(c) for c in confidences2]
                }
            }
        }

    return results

def save_comparison(results: Dict, output_path: str):
    with open(output_path, 'w') as f:
        json.dump(results, f, indent=2)

def calculate_aggregate_stats(comparison_results: Dict) -> Dict:
    conf_diffs = []
    count_diffs = []

    for file_result in comparison_results.values():
        conf_diffs.append(file_result['confidence_diff'])
        count_diffs.append(file_result['detection_count_diff'])

    return {
        'mean_confidence_diff': safe_mean(conf_diffs),
        'mean_count_diff': safe_mean(count_diffs),
        'total_files_compared': len(comparison_results)
    }
```

```python
metrics1 = load_metrics('/content/output_ffa/detections/gt_its.json')
metrics2 = load_metrics('/content/output_ffa/detections/ffa_its.json')

comparison_results = compare_detections(metrics1, metrics2)
save_comparison(comparison_results, '/content/delta_its.json')

stats = calculate_aggregate_stats(comparison_results)
print("\nAggregate Statistics:")
print(f"- Average confidence difference: {stats['mean_confidence_diff']:.4f}")
print(f"- Average detection count difference: {stats['mean_count_diff']:.4f}")
print(f"- Total files compared: {stats['total_files_compared']}")
```

```python
import json
from collections import defaultdict

def analyze_json_files(file_paths: List[str]):
    for path in file_paths:
        try:
            with open(path) as f:
                data = json.load(f)

            counters = {
                'total_files': 0,
                'zero_confidences': 0,
                'zero_names': 0,
                'zero_both': 0,
                'missing_detections': 0
            }

            zero_conf_files = []
            zero_name_files = []
            zero_both_files = []
            missing_det_files = []

            for i, filename in enumerate(data['filenames']):
                counters['total_files'] += 1

                try:
                    det = data['detections'][i]

                    confidences = det.get('confidences', [])
                    conf_empty = len(confidences) == 0

                    names = det.get('names', [])
                    names_empty = len(names) == 0

                    if conf_empty and names_empty:
                        counters['zero_both'] += 1
                        zero_both_files.append(filename)
                    elif conf_empty:
                        counters['zero_confidences'] += 1
                        zero_conf_files.append(filename)
                    elif names_empty:
                        counters['zero_names'] += 1
                        zero_name_files.append(filename)

                except (KeyError, IndexError):
                    counters['missing_detections'] += 1
                    missing_det_files.append(filename)

            print(f"\nSummary for {path}:")
            print(f"Total files: {counters['total_files']}")
            print(f"Files with zero-length confidences: {counters['zero_confidences']}")
            print(f"Files with zero-length names: {counters['zero_names']}")
            print(f"Files with both empty: {counters['zero_both']}")
            print(f"Files missing detections: {counters['missing_detections']}")

            def print_file_list(title, files, max_display=5):
                if files:
                    print(f"\n{title}:")
                    for f in files[:max_display]:
                        print(f"  - {f}")
                    if len(files) > max_display:
                        print(f"  ... and {len(files)-max_display} more")

            print_file_list("Files with ZERO confidences", zero_conf_files)
            print_file_list("Files with ZERO names", zero_name_files)
            print_file_list("Files with BOTH empty", zero_both_files)
            print_file_list("Files MISSING detections", missing_det_files)

            if zero_conf_files or zero_name_files or zero_both_files or missing_det_files:
                print("\nExample JSON structure for first problematic file:")
                example_file = (zero_conf_files + zero_name_files + zero_both_files + missing_det_files)[0]
                idx = data['filenames'].index(example_file)
                print(json.dumps({
                    "filename": example_file,
                    "detection_data": data['detections'][idx] if idx < len(data['detections']) else "MISSING"
                }, indent=2))

        except Exception as e:
            print(f"\nError analyzing {path}: {str(e)}")

if __name__ == '__main__':
    # json_files = ['/content/output_ffa/detections/gt_samplees.json', '/content/output_ffa/detections/ffa_samplees.json']
    json_files = ['/content/gt_its.json', '/content/ffa_its.json']
    analyze_json_files(json_files)
```

## Plotting the detection differences

```python
import matplotlib.pyplot as plt
import numpy as np
from typing import Dict

def plot_detection_differences(comparison_results: Dict, first_n: int = 50):
    first_results = dict(list(comparison_results.items())[:first_n])

    filenames = list(first_results.keys())
    conf_diffs = [res['confidence_diff'] for res in first_results.values()]
    count_diffs = [res['detection_count_diff'] for res in first_results.values()]

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))
    fig.suptitle(f'Detection Comparison (First {first_n} Images)', fontsize=14)

    ax1.bar(filenames, conf_diffs, color='skyblue')
    ax1.set_ylabel('Confidence Difference\n(Set1 - Set2)', fontsize=10)
    ax1.axhline(y=0, color='gray', linestyle='--')
    ax1.set_xticks([])
    ax1.set_title('Average Confidence Difference per Image', fontsize=12)

    ax2.bar(filenames, count_diffs, color='salmon')
    ax2.set_ylabel('Detection Count Difference\n(Set1 - Set2)', fontsize=10)
    ax2.axhline(y=0, color='gray', linestyle='--')
    ax2.set_xticks([])
    ax2.set_title('Detection Count Difference per Image', fontsize=12)

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    with open('/content/delta_its.json') as f:
        comparison_results = json.load(f)

    plot_detection_differences(comparison_results, first_n=50)
```

# Downloading images

```python
import os
from google.colab import files
from IPython.display import display, Javascript
from PIL import Image
import zipfile

def download_images(folder_path: str, output_zip: str = 'images.zip'):

    valid_extensions = ('.jpg', '.jpeg', '.png')

    image_files = [
        f for f in os.listdir(folder_path)
        if os.path.splitext(f)[1].lower() in valid_extensions
    ]

    if not image_files:
        print("No images found in the specified folder.")
        return

    print(f"Found {len(image_files)} images.")

    with zipfile.ZipFile(output_zip, 'w') as zipf:
        for img_file in image_files:
            img_path = os.path.join(folder_path, img_file)

            try:
                with Image.open(img_path) as img:
                    img.verify()
                zipf.write(img_path, img_file)
            except (IOError, SyntaxError) as e:
                print(f"Skipping invalid image file: {img_file} ({str(e)})")

    files.download(output_zip)
    print(f"Download initiated for {output_zip} with {len(image_files)} images.")
```
