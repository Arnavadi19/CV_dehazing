```python
from PIL import Image
import os
import glob

# Set the directory path
directory = "/home/arnav/CVproj/D4/checkpoints/video20"

# Get all jpg files in the directory
jpg_files = glob.glob(os.path.join(directory, "*.jpg"))

# Convert each file
for jpg_file in jpg_files:
    img = Image.open(jpg_file)
    png_file = os.path.splitext(jpg_file)[0] + ".png"
    img.save(png_file)
    print(f"Converted {os.path.basename(jpg_file)} to {os.path.basename(png_file)}")

print(f"Converted {len(jpg_files)} images from JPG to PNG")

#Optional: Remove original jpg files
for jpg_file in jpg_files:
    os.remove(jpg_file)
```

```python
import cv2
import os
from natsort import natsorted

# --- Input parameters ---
image_folder = "/home/arnav/CVproj/D4/checkpoints/video20/results/reconstruct"         # Folder with .png images
output_video = "/home/arnav/CVproj/D4/checkpoints/video20/results/reconstruct/output_video.mp4"   # Output video filename
fps = 10                            # Frames per second

# --- Get sorted list of image paths ---
images = [img for img in os.listdir(image_folder) if img.endswith(".png")]
images = natsorted(images)  # Sort naturally: frame_1.png, frame_2.png, ...

if not images:
    raise ValueError("No PNG images found in the folder.")

# --- Get frame size from the first image ---
first_frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, _ = first_frame.shape

# --- Initialize video writer ---
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4
video_writer = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

# --- Write images to video ---
for img_name in images:
    frame = cv2.imread(os.path.join(image_folder, img_name))
    video_writer.write(frame)

video_writer.release()
print(f"Video saved to: {output_video}")
```
