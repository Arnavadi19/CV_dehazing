```python
#!/usr/bin/env python3
import os
import re

def rename_files(directory):
    """Rename files by removing the underscore and number before extension"""
    # Get all PNG files
    files = [f for f in os.listdir(directory) if f.endswith('.png')]
    
    # Track stats
    renamed_count = 0
    skipped_count = 0
    
    # Process each file
    for filename in files:
        # Use regex to match the pattern and extract the base name
        match = re.match(r'(\d+)_\d+\.png', filename)
        if match:
            base_name = match.group(1)
            new_filename = f"{base_name}"
            old_path = os.path.join(directory, filename)
            new_path = os.path.join(directory, new_filename)
            
            # Check if destination already exists
            if os.path.exists(new_path):
                print(f"SKIPPED: {new_filename} already exists")
                skipped_count += 1
                continue
                
            # Rename the file
            os.rename(old_path, new_path)
            print(f"Renamed: {filename}  {new_filename}")
            renamed_count += 1
    
    print(f"\nComplete! Renamed {renamed_count} files, skipped {skipped_count} files")

# Directories to process
directories = [
    "/home/arnav/CVproj/RESIDE-6K/test/GT_indoor",
    "/home/arnav/CVproj/RESIDE-6K/test/hazy_indoor"
]

# Run the renaming process for each directory
for directory in directories:
    print(f"\nProcessing directory: {directory}")
    if os.path.exists(directory):
        rename_files(directory)
    else:
        print(f"ERROR: Directory not found: {directory}")
```
